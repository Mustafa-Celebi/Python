```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cars.csv")
df
```

```python
seat_column = df["Seats"]
value = seat_column.value_counts().head(4)

# Pasta grafiÄŸi oluÅŸturma
plt.figure(figsize=(8, 8))
plt.pie(value, labels=value.index, colors=sns.color_palette("pastel"), autopct='%1.1f%%', startangle=140)

# GrafiÄŸi gÃ¶sterme
plt.title('Distribution of Average Number of Seats')
plt.show()
```

```python
seat = df['Seats'].value_counts().sort_index()

year = (seat.index * seat).sum() / seat.sum()

seat.plot(kind='bar', figsize=(13, 6))
sns.barplot(x=seat.index, y=seat.values, palette='magma')
plt.xlabel('Seat Counts')
plt.ylabel('Car Counts')
plt.title('Average Seat Counts')
plt.show()
```
